```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots(1, 1)
x = np.linspace(norm.ppf(.0001), norm.ppf(.9999), 1000)
ax.plot(x,norm.pdf(x), label='standard normal pdf', lw=2, alpha=.6)
rv = norm.rvs(size=1000)
ax.hist(rv, normed=True, alpha=.2)
ax.legend(loc='upper right')
plt.show()
```
